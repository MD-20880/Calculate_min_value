Define Functions
=========
Calculate() function go through every possible value in step within given range. Calculate its simulation result and find the one with smallest value

```python
def calculate(startx,starty,endx,endy,func,linspace=10):
    x = np.linspace(startx,endx,num = linspace)
    y = np.linspace(starty,endy,num = linspace)
    xs,ys = np.meshgrid(x,y)
    zs = [[func(xs[i][j] , ys[i][j])for j in range(len(x))] for i in range(len(y))]
    meshes = np.dstack((xs,ys,zs))
    new_mesh = meshes.reshape(meshes.shape[1]*meshes.shape[0],meshes.shape[2])
    min_index = np.argmin(new_mesh,axis=0)[2]


    next_x_start = max(x[x<new_mesh[min_index][0]],default=new_mesh[min_index][0])
    next_x_end = min(x[x>new_mesh[min_index][0]],default=new_mesh[min_index][0])
    next_y_start = max(y[y<new_mesh[min_index][1]],default=new_mesh[min_index][1])
    next_y_end = min(y[y>new_mesh[min_index][1]],default=new_mesh[min_index][1])

    return next_x_start,next_y_start,next_x_end,next_y_end,new_mesh[min_index]
```

```python
def loop_calculate(startx,starty,endx,endy,func,linspace=10,looptimes = 5):
    for i in range(looptimes):
        print("xs = {0}\nys = {1}\nxe = {2}\nye = {3}".format(startx,starty,endx,endy))
        startx,starty,endx,endy,result = calculate(startx,starty,endx,endy,func,linspace=linspace)
        print(result)
    return result

```
